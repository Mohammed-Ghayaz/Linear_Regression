**Linear Regression using Gradient Descent**

```python
#Uses the linear equation of the regression line to predict the output
def predict(X, theta, b):
    return X.dot(theta) + b

#Loss function
def mean_sum_of_squares_error(X, y, theta, b):
    m = len(y)
    prediction = predict(X, theta, b)
    loss = (1/(2*m))*np.sum((y - prediction)**2)
    return loss


#Applying Gradient Descent to update the weights and bias
def gradient_descent(X, y, theta, b):
    m = len(y)

    grad_theta = np.zeros(len(theta))
    gradient_b = 0

    pred = predict(X, theta, b)

    gradient_b = -(1/m) * np.sum(y - pred)
    for j in range(len(theta)):
        grad_theta[j] = -(1/m) * np.sum((y - pred)*X.iloc[:, j])
        
    b = b - 0.01*gradient_b
    for j in range(len(theta)):
        theta[j] = theta[j] - 0.01*grad_theta[j]

    return theta, b

def linear_regression(X, y):
    theta = np.zeros(X.shape[1])
    b = 0
    loss_val = []
    for i in range(5000):
        theta, b = gradient_descent(X, y, theta, b)
        if i%100==0:
            loss = mean_sum_of_squares_error(X, y, theta, b)
            loss_val.append(f"Cost at iteration {i}: {loss}")
        
    return theta, loss_val, b
        
    
```

```python
theta, loss_history, b = linear_regression(X, y)
print("Optimized theta:", theta) #Prints the weights for each feature
print("Optimized bias (b):", b)  
for loss in loss_history:
    print(loss)
```
